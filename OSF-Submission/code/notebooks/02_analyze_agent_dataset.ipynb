# 02 — Analyze ASRA v0.1 Agent Dataset

Phase 1 QA: inspect exported transitions, episode summaries, and state graph stats.

```python
import json
from pathlib import Path

import pandas as pd

ROOT = Path('..').resolve()
EXPORTS = ROOT / 'data' / 'exports'
JSONL = EXPORTS / 'asra_v0_1_transitions.jsonl'
SUMMARY = EXPORTS / 'asra_v0_1_episode_summary.csv'
GRAPH = EXPORTS / 'asra_v0_1_state_graph.json'
```

```python
rows = [json.loads(line) for line in JSONL.read_text().splitlines() if line.strip()]
df = pd.DataFrame(rows)
print('Transitions:', len(df))
print('Episodes:', df['episode_id'].nunique())
print('Unique state hashes:', df['state_hash'].nunique())
df['action'].apply(lambda a: a['name']).value_counts().head(10)
```

```python
summary = pd.read_csv(SUMMARY)
display(summary)
print('Terminal outcomes:', summary['final_status'].value_counts().to_dict())
```

```python
graph = json.loads(GRAPH.read_text())
print('Graph nodes:', len(graph['nodes']))
print('Graph edges:', len(graph['edges']))
reports = list((ROOT / 'data' / 'analysis' / 'action_reports').glob('*.json'))
print('Action test reports:', len(reports))
```
