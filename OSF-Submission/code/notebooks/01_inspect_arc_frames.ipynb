# 01 — Inspect ARC-AGI-3 Frames

Phase 1 notebook: parse and visualize normalized frames from mock, replay, or raw JSON.

```python
import json
from pathlib import Path

from asra.env.frame_parser import parse_frame
from asra.env.replay_backend import ReplayArcAGI3Environment
from asra.env.arc_agi3_runner import MockArcAGI3Environment

ROOT = Path('..').resolve()
RAW = ROOT / 'data' / 'raw' / 'sample_arc_agi3_replay.json'
```

```python
# Mock frame
mock_raw = MockArcAGI3Environment().reset('mock-game', 'mock-level')
mock_frame = parse_frame(mock_raw)
print('Mock:', mock_frame.game_id, mock_frame.status, mock_frame.height, 'x', mock_frame.width)
for row in mock_frame.grid:
    print(row)
```

```python
# Replay file frame (offline ARC-AGI-3 style)
if RAW.is_file():
    replay_env = ReplayArcAGI3Environment(RAW)
    replay_frame = parse_frame(replay_env.reset())
    print('Replay:', replay_frame.game_id, replay_frame.level_id)
    for row in replay_frame.grid:
        print(row)
else:
    print('Run: python scripts/generate_sample_replay.py')
```

```python
# Optional: load any raw JSON frame dump
raw_dir = ROOT / 'data' / 'raw'
for path in sorted(raw_dir.glob('*.json')):
    payload = json.loads(path.read_text())
    print(path.name, 'keys:', list(payload.keys())[:6])
```
